```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Position' : [1,2,3,4,5],
    'Club' : ["Liverpool","chelsea","Arsenal","Man City","Man United"],
    'Point' : [40,38,35,35,28],
    'Goals' : [200,400,231,22,345]
})

df
```

### 1. Sorting by index labels

```python
df.sort_index(ascending=False)
# this will sort in descending order by index
```

```python
df.sort_index(ascending=True)
# this will sort in ascending order by index
```

```python
df.sort_index(axis=1)
# this will sort by column
```

```python
df.sort_index(axis=0)
# this will sort by column
```

### 2. sorting by column values

```python
df.sort_values(by='Goals')
# this will sort by column goals
```

```python
df.sort_values('Goals')
# this will sort by column goals
```

```python
df['Goals'].sort_values(ascending=False)
# this will sort column goals by descending order
```

```python
df['Goals'].sort_values(ascending=True)
# this will sort column goals by ascending order
```

### 3. sorting by combination of both

```python
df.set_index('Goals').head()
# this will set goals column as index
```

```python
df.sort_values(by=['Goals','Position'])
# this will sort by index (goals) and by column (postion)
```

```python
df.sort_values(by=['Goals','Club'])
# this will sort by index (goals) and by column (Club)
```
